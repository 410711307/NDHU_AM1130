```python
from scipy.ndimage import gaussian_filter
def blur(img, dpi):
    return gaussian_filter(img, (3,3,0)), -0, 0
```

```python
plt.figure(figsize=(5,5))
ax = plt.gca()
ax.set_facecolor('black')

t = np.linspace(0, 2*np.pi, 101)
plt.plot(np.cos(t), np.sin(t), 'w--')
plt.plot(.7*np.cos(t), .7*np.sin(t), 'w', linewidth=15, agg_filter=blur)
plt.plot(.7*np.cos(t), .7*np.sin(t), 'w')

plt.plot(1.3*np.cos(t), 1.3*np.sin(t), 'w')

TEXT = 'NDHUγAMλ'
for i in range(len(TEXT)):
    s = np.pi*2*i/len(TEXT)
    x0,y0 = 1.15*np.cos(s), 1.15*np.sin(s)
    ax.text(x0, y0, TEXT[i], 
            color='white', va='center', ha='center', 
            rotation=s*180/np.pi-90, fontsize=16)
    plt.plot(.15*np.cos(t)+x0, .15*np.sin(t)+y0, 'w')
    

t = np.linspace(0, 4*np.pi, 6)
plt.plot(.7*np.cos(t+np.pi/10), .7*np.sin(t+np.pi/10), 'w')


t = np.linspace(0, 2*np.pi, 5)
plt.plot(np.cos(t), np.sin(t), 'w:')
plt.plot(np.cos(t+np.pi/4), np.sin(t+np.pi/4), 'w:')
```

```python
def plotx(*a, linewidth=1, **k):
    plt.plot(*a, **k, linewidth=15*linewidth, alpha=0.3, agg_filter=blur)
    plt.plot(*a, **k, linewidth=7*linewidth, alpha=0.8, agg_filter=blur)
    plt.plot(*a, **k, linewidth=linewidth)
```

```python
from scipy.ndimage import gaussian_filter
def Blur(sigma):
    def blur(img, dpi):
        return gaussian_filter(img, (sigma,sigma,0)), -0, 0
    return blur
```

```python
def plotx(*a, linewidth=2, alpha=1.0, **k):
    plt.plot(*a, **k, linewidth=12*linewidth, alpha=alpha*0.3, agg_filter=Blur(5))
    plt.plot(*a, **k, linewidth=5*linewidth, alpha=alpha*0.6, agg_filter=Blur(3))
    plt.plot(*a, **k, linewidth=linewidth, alpha=alpha)
```

```python
import matplotlib.font_manager

plt.figure(figsize=(10,10))
ax = plt.gca()
ax.set_facecolor('black')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

t = np.linspace(0, 2*np.pi, 101)
for d in range(10):
    shift = np.random.uniform(low=0, high=2*np.pi)
    noise = np.random.uniform(low=-0.05, high=0.05, size=t.shape)
    freq = np.random.uniform(low=0.5, high=2)
    r = np.random.uniform(low=0.5, high=2, size=t.shape)
    color = tuple(np.random.uniform(0.8,1, size=3))
    z = 1.33+.05*r*np.sin(freq*t+shift+noise)+noise
    plt.plot(z*np.cos(t), z*np.sin(t), c=color, linewidth=12,
             alpha=0.1, agg_filter=Blur(7))
    plt.plot(z*np.cos(t), z*np.sin(t), c=color, linewidth=7, alpha=0.3, 
             agg_filter=Blur(5))
    plt.plot(z*np.cos(t), z*np.sin(t), c=color, linewidth=2, alpha=.5, 
             agg_filter=Blur(1))

color = (.8,1,1)
t = np.linspace(0, 2*np.pi, 5)
plotx(np.cos(t), np.sin(t), color=color, alpha=0.6)
plotx(np.cos(t+np.pi/4), np.sin(t+np.pi/4), color='w', alpha=0.6)

t = np.linspace(0, 2*np.pi, 101)
plt.plot(.76*np.cos(t), .76*np.sin(t), linewidth=14, color='k')
r = 1.25+0.3*np.sin(31*t)**2
plt.plot( r*np.cos(t), r*np.sin(t),  color=(1,1, 1), 
         linewidth=20,alpha=0.3, agg_filter=Blur(8))
plotx(np.cos(t), np.sin(t), linestyle='--', marker='X', markersize=4, color=color, linewidth=1)
plotx(.7*np.cos(t), .7*np.sin(t), color=color)
plotx(.82*np.cos(t), .82*np.sin(t), color=color)


plotx(1.3*np.cos(t), 1.3*np.sin(t), color=color)

TEXT = 'NDHUMATH'
font0 = matplotlib.font_manager.FontProperties(family='Lobster Two', 
            size=42, style='italic', weight='heavy')
font = matplotlib.font_manager.FontProperties(family='Lobster Two', 
            size=36, style='italic', weight='light')
for i in range(len(TEXT)):
    s = np.pi*2*(.5-i/len(TEXT)) #-np.pi/8
    x0,y0 = 1.11*np.cos(s), 1.13*np.sin(s)
    ax.text(x0, y0, TEXT[i], 
            color=color, va='center', ha='center', 
            rotation=s*180/np.pi-90, fontproperties=font0, agg_filter=Blur(3))
    ax.text(x0, y0, TEXT[i], 
            color='w', va='center', ha='center', 
            rotation=s*180/np.pi-90, fontproperties=font, alpha=.8)
    x0,y0 = 1.15*np.cos(s), 1.15*np.sin(s)
    plotx(.15*np.cos(t)+x0, .15*np.sin(t)+y0, color=color)
    

TEXT = 'Department of Applied Mathematics, National Dong Hwa University, Hualien, Taiwan'
font = matplotlib.font_manager.FontProperties(family='Lobster Two', 
            size=15, style='italic', weight='light')
for i in range(len(TEXT)):
    s = np.pi*2*(.5-i/len(TEXT))
    x0,y0 = .75*np.cos(s), .75*np.sin(s)
    ax.text(x0, y0, TEXT[i], 
            color='w', va='center', ha='center', 
            rotation=s*180/np.pi-90, fontproperties=font, alpha=.7)

    
t = np.linspace(0, 4*np.pi, 6)
plotx(.7*np.cos(t+np.pi/10), .7*np.sin(t+np.pi/10), color=color)



```
