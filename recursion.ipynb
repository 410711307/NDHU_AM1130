```
def f(n):
  print(n)
  if n == 1:
    print("done")
  else:
    f(n-1)
```

```
def f(n):
  #print("calling f", n )
  if n == 0:
    return 1
  elif n == 1:
    return 1
  else:
    return f(n-1)+f(n-2)
```

```
for i in range(10):
  print(i, '->', f(i))
```

```
%%timeit
f(35)
```

```
fib = [1,1]
for i in range(2, 36):
  x=fib[-1]+fib[-2]
  print(i, '-->', x)
  fib.append(x)
```

```
def f(n, mem={}):
  print(n , mem)
  if n not in mem:    
    if n == 0:
      r = 1
    elif n == 1:
      r =  1
    else:
      r = f(n-1)+f(n-2)
    mem[n] = r
  return mem[n]
 
```

```
def f(n, mem={0:1, 1:1}):
  if n not in mem:    
    mem[n] = f(n-1)+f(n-2)    
  return mem[n]
 
```

```
f(35)
```

```
from functools import lru_cache

@lru_cache(maxsize=1024)
def f(n):
  if n == 0:
    return 1
  elif n == 1:
    return 1
  else:
    return f(n-1)+f(n-2)
```
