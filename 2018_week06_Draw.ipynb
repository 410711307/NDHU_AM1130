```
import numpy as np
img = np.zeros( (600, 800, 3), dtype=np.float)
```

```
#%matplotlib inline
import matplotlib.pyplot as plt
```

```
plt.imshow(img)
```

```
img = np.zeros( (600, 800, 3), dtype=np.float)
for y in range(0,600, 100):
  for x in range(0, 800, 100):
    if (x+y)%200 == 100:
      img[y:y+100,x:x+100]=(0.5, 0.1, 1)
plt.imshow(img)
```

```
img = np.zeros( (600, 800, 3), dtype=np.float)
img[:, 350:450]=(1,0,0)
img[150:250, 150:650]=(1,0,0)
plt.imshow(img)
```

```
img = np.zeros( (600, 800, 3), dtype=np.float)
for y in range(600):
  for x in range(800):
    if (x-400)**2 + (y-300)**2 < 200**2:
      img[y,x]=1
plt.imshow(img)
```
