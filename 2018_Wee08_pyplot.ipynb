```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```
x = np.linspace(0, 2*np.pi, 1000)
```

```
plt.plot(x, np.sin(x))
```

```
plt.plot(x, 0.1*x**3-0.2*x**2-x+1)
```

```
plt.plot(np.cos(x), np.sin(x))
```

```
plt.figure(figsize=(5,5))
t=np.linspace(0, 2*np.pi, 6)
plt.plot(np.cos(t), np.sin(t))
```
