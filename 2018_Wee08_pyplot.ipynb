```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```
x = np.linspace(0, 2*np.pi, 1000)
```

```
plt.plot(x, np.sin(x))
```

```
plt.plot(x, 0.1*x**3-0.2*x**2-x+1)
```

```
plt.plot(np.cos(x), np.sin(x))
```

```
plt.figure(figsize=(5,5))
t=np.linspace(0, 2*np.pi, 6)
plt.plot(np.cos(t+0.1*np.pi), np.sin(t+0.1*np.pi))
```

```
plt.figure(figsize=(5,5))
t=np.linspace(np.pi/2, np.pi/2+2*np.pi, 6)
plt.plot(np.cos(t), np.sin(t))
```

```
plt.figure(figsize=(5,5))
t=np.linspace(np.pi/2, np.pi/2+2*np.pi, 12)
plt.plot(np.cos(t), np.sin(t))
```

```
plt.figure(figsize=(5,5))
t=np.linspace(np.pi/2, np.pi/2+4*np.pi, 6)
plt.plot(np.cos(t), np.sin(t))
```

```
plt.figure(figsize=(5,5))
t=np.linspace(np.pi/2, np.pi/2+1*2*np.pi, 13)
plt.plot(np.cos(t), np.sin(t))
```

```
plt.figure(figsize=(5,5))
t=np.linspace(np.pi/2, np.pi/2+2*np.pi, 11)
r = np.array([1, .3] * 5 + [1])
plt.plot(np.cos(t)*r, np.sin(t)*r)
```
