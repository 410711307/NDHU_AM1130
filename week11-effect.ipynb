```python
from scipy.ndimage import gaussian_filter
def Blur(sigma):
    def blur(img, dpi):
        return gaussian_filter(img, (sigma,sigma,0)), -0, 0
    return blur
```

```python
plt.figure(figsize=(5,5))
ax = plt.gca()
ax.set_facecolor('black')

t = np.linspace(0, 2*np.pi, 101)
plt.plot(np.cos(t), np.sin(t), 'w', linewidth=10, agg_filter=Blur(5))
```

```python
plt.figure(figsize=(8,5))
ax = plt.gca()
ax.set_facecolor('black')
plt.xlim(0,7)
plt.ylim(-1, 1)

t = np.linspace(0, 7, 101)
for d in range(20):
    shift = np.random.uniform(low=0, high=2*np.pi)
    noise = np.random.uniform(low=-0.05, high=0.05, size=t.shape)
    freq = np.random.uniform(low=0.5, high=2)
    r = np.random.uniform(low=0.5, high=2, size=t.shape)
    color = tuple(np.random.uniform(0.8,1, size=3))
    plt.plot(t, .2*r*np.sin(freq*t+shift), c=color, linewidth=5, alpha=0.3, agg_filter=Blur(7))
    plt.plot(t, .2*r*np.sin(freq*t+shift)+noise, c=color, linewidth=5, alpha=0.5, agg_filter=Blur(3))
    plt.plot(t, .2*r*np.sin(freq*t+shift+noise)+noise, c=color, linewidth=1, alpha=.7, agg_filter=Blur(1))
```
