```python
import numpy as np
from PIL import Image
```

```python
a = np.zeros((600,800,3), dtype='uint8')
Image.fromarray(a)
```

```python
def line(x1, y1, x2, y2, c, w=3):
    k = max(abs(x2-x1), abs(y2-y1))
    for i in range(k+1):
        x = int(x1 * i/k  + x2 *(1-i/k))
        y = int(y1 * i/k  + y2 *(1-i/k))
        a[y-w:y+w, x-w:x+w] = c
        
```

```python
def circle(cx, cy, r, c):
    for x in range(800):
        for y in range(600):
            if (x-cx)**2 + (y-cy)**2 < r**2:
                a[y,x]=c
```

```python
def path(xylist, c):
    l = len(xylist)
    for i in range(l-1):
        x1, y1 = xylist[i]
        x2, y2 = xylist[i+1]
        line(x1, y1, x2, y2, c)
```

```python
a[:] = 0
# 禾
line(100,100, 50,140, 255)
line(40,150, 120, 150,  255)
line(75, 120, 75, 200, 255)
line(75, 150, 40, 200, 255)
line(75, 150, 120, 200, 255)
# 女
c = (0,255,0) # 綠色
path([(75,210), (50,240), (75,270) ], c)
line(85,240, 50,270, c)
line(40,240, 100, 240, c)
# 鬼
x, y =  160,130 # 田的左上角
c = (0,0,255) # 藍色 
path([(x+40,y-30) , (x,y), 
      (x,y+50), (x+70, y+50), (x+70,y),(x,y) ], c)
line(x, y+25, x+70, y+25, c)
path( [(x+35, y), (x+35, y+50), (x-10, y+150)], c)
path( [(x+35, y+50), (x+50, y+150), (x+90, y+150), (x+90, y+135)], c)
path([(x+70, y+115), (x+60, y+135), (x+80, y+135), (x+75, y+130), (x+85, y+140)], c)
Image.fromarray(a)
```
